```python
def calc_wind_mag_from_ds(ds):
    u = ds['u']
    v = ds['v']
    u = u.interp(xh=v.xf.values)
    v = v.interp(yh=u.yf.values)
    # Dims
    u = u.rename({'xh':'x', 'yf':'y'})
    v = v.rename({'xf':'x', 'yh':'y'})
    return np.sqrt(u**2 + v**2)
```

```python
import glob
def calc_wind_mag(ds):
    ne = len(list(ds.data_vars)) // 2 
    assert len(list(ds.data_vars)) % 2 == 0, "Weird number of data-arrays, check input ds"
    for i in range(ne):
        ds[f'M_memb{i}'] = np.sqrt(ds[f'u_memb{i}']**2 + ds[f'v_memb{i}']**2)
    return ds


def load_assimilation_step(step, data_dir, data='data'):
    
    if data=='data':
        if step=='posterior':
            print("Loading data posterior")
            dirs = sorted(glob.glob(f"{data_dir}/data_posterior_step*_member*.nc"))
        else:
            print(f"Loading data at step {step}")
            dirs = sorted(glob.glob(f"{data_dir}/data_prior_step{str(step).zfill(2)}_member*.nc"))
    elif data=='model':
        print(f"Loading model at step {step}")
        dirs = sorted(glob.glob(f"{data_dir}/model_post_step{str(step).zfill(2)}_member*.nc"))       
    
    ds_all = xr.Dataset()
    for i,dir in enumerate(dirs):
        ds = xr.open_dataset(dir)
        # Round time to minutes
        if data=='data':
            ds['time'] = ds['time'].dt.round('min')
            ds_all[f'u_memb{i}'] = ds['u_predobs']
            ds_all[f'v_memb{i}'] = ds['v_predobs']
        else:
            ds_all[f'u_memb{i}'] = ds['u']
            ds_all[f'v_memb{i}'] = ds['v']
            ds_all[f'thl_memb{i}'] = ds['Thl']
            ds_all[f'qt_memb{i}']  = ds['qt']
            ds_all[f'w_memb{i}']   = ds['w']
    
    return ds_all



def compute_ens_mean(ds):
    u_mean, v_mean, M_mean = 0, 0, 0
    thl_mean, qt_mean, w_mean = 0, 0, 0
    idx_x, idx_y, idx_M = 0, 0, 0
    
    for i, var in enumerate(ds.data_vars):
        da = ds[var]
        if 'time' in ds.sizes:
            da['time'] = da['time'].dt.round('min')
        if 'u' in var:
            u_mean += da
            idx_x += 1
        elif 'v' in var:
            v_mean += da
            idx_y += 1
        elif 'M' in var:
            M_mean += da
            idx_M += 1
        elif 'thl' in var:
            thl_mean += da
        elif 'qt' in var:
            qt_mean += da
        elif 'w' in var:
            w_mean += da
        else:
            print(f"Different DataArray {var}")
    
    assert idx_x == idx_y, "Not same amount of u and v variables"
    assert idx_x != 0, "DataSet contains no variables u or v!"
    
    u_mean /= idx_x
    v_mean /= idx_y
    
    thl_mean /= idx_y
    qt_mean /= idx_y
    w_mean /= idx_y

    if idx_M != 0:
        M_mean /= idx_M
    
    ds_mean = xr.Dataset({
        'u_mean': u_mean,
        'v_mean': v_mean,
        'M_mean': M_mean,
        'thl_mean': thl_mean,
        'qt_mean': qt_mean,
        'w_mean': w_mean,
        })
    return ds_mean
```

```python
# Process data assiimilation data
ds_prior     = load_assimilation_step(step=0, data_dir=data_dir)
ds_posterior = load_assimilation_step(step='posterior', data_dir=data_dir)
ds_prior     = calc_wind_mag(ds=ds_prior)
ds_posterior = calc_wind_mag(ds=ds_posterior)
ds_prior_mean     = compute_ens_mean(ds_prior)
ds_posterior_mean = compute_ens_mean(ds_posterior)


# Process data assiimilation model
ds_model_step0 = load_assimilation_step(step=0, data_dir=data_dir, data='model')
ds_model_post  = load_assimilation_step(step=3, data_dir=data_dir, data='model')
ds_model_prior_mean = compute_ens_mean(ds_model_step0)
ds_model_post_mean  = compute_ens_mean(ds_model_post)

# Change dataset datasets
ds_posterior_mean = ds_posterior_mean.rename({'u_mean':'u', 'v_mean':'v', 'M_mean':'M'})
ds_prior_mean = ds_prior_mean.rename({'u_mean':'u', 'v_mean':'v', 'M_mean':'M'})

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_combined_hist_with_gaussian_fit(ax, data1, data2, label1="Data 1", label2="Data 2", color1="steelblue", color2="darkorange"):
    """
    Plots histograms of two datasets with fitted Gaussian curves on a given axis.
    
    Parameters:
    - ax: matplotlib axis on which to plot
    - data1: flattened data for the first dataset
    - data2: flattened data for the second dataset
    - label1: label for the first dataset
    - label2: label for the second dataset
    - color1: color for the first dataset
    - color2: color for the second dataset
    """
    # Calculate Gaussian parameters for each dataset
    mean1, std1 = np.mean(data1), np.std(data1)
    mean2, std2 = np.mean(data2), np.std(data2)
    
    # Plot histograms
    ax.hist(data1, bins=250, density=True, histtype='step', linewidth=3, label=f'{label1}', color=color1, alpha=0.9)
    ax.hist(data2, bins=250, density=True, histtype='step', linewidth=3, label=f'{label2}', color=color2, alpha=0.9)
    
    # Plot Gaussian fits
    x1 = np.linspace(mean1 - 4 * std1, mean1 + 4 * std1, 1000)
    x2 = np.linspace(mean2 - 4 * std2, mean2 + 4 * std2, 1000)
    ax.plot(x1, norm.pdf(x1, mean1, std1), color=color1, alpha=0.5, linestyle='--', linewidth=2, label=f'Gaussian fit {label1}')
    ax.plot(x2, norm.pdf(x2, mean2, std2), color=color2, alpha=0.5, linestyle='--', linewidth=2, label=f'Gaussian fit {label2}')
    
    # Vertical lines at means
    ax.axvline(mean1, color=color1, linestyle='-', linewidth=1.5, alpha=0.25)
    ax.axvline(mean2, color=color2, linestyle='-', linewidth=1.5, alpha=0.25)
    
    # Aesthetic adjustments
    ax.set_xlabel("Value", fontsize=12)
    ax.set_ylabel("Density", fontsize=12)
    ax.legend(loc='upper left', fontsize=10, frameon=False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)
    ax.grid(alpha=0.25)

# Flatten and define data for u and v distributions
u_prior = ds_model_prior_mean['u_mean'].sel(zf=slice(0, 2000)).values.flatten()
u_post = ds_model_post_mean['u_mean'].sel(zf=slice(0, 2000)).values.flatten()
v_prior = ds_model_prior_mean['v_mean'].sel(zf=slice(0, 2000)).values.flatten()
v_post = ds_model_post_mean['v_mean'].sel(zf=slice(0, 2000)).values.flatten()

# Create figure and two subplots side by side
fig, axs = plt.subplots(ncols=2, figsize=(12, 4), sharey=True, constrained_layout=True)

# Plot combined prior distributions for u and v
plot_combined_hist_with_gaussian_fit(axs[0], u_prior, v_prior, label1="u prior", label2="v prior", color1="steelblue", color2="darkorange")
axs[0].set_title("Prior Distributions", fontsize=12)

# Plot combined post distributions for u and v
plot_combined_hist_with_gaussian_fit(axs[1], u_post, v_post, label1="u post", label2="v post", color1="steelblue", color2="darkorange")
axs[1].set_title("Post Distributions", fontsize=12)

plt.show()
```
